```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#used for displaying plots easily: below code
%matplotlib inline 
```

Now you need libraries to open the url and then to extract data from the html page


```python
from urllib.request import urlopen #for opening url
from bs4 import BeautifulSoup  #for extracting data from the html page
```

```python
#Opening the url

url ='http://www.hubertiming.com/results/2017GPTR10K'
html = urlopen(url)
```

```python
#Creating a beautiful soup object

soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
#Getting info from the webpage
#fetching title

title = soup.title
print (title)
```

```python
#Printing the text
text = soup.get_text()
print(text)
```

To extract specific tags use find_all method

 < a > for hyperlinks, < table > for tables, < tr > for table rows, < th > for table headers, and < td > for table cells

```python
#to extract all the a tags on the page

soup.find_all('a')
```

```python
#To extract just the hyperlinks
link_tags = soup.find_all('a')
for link in link_tags:
    print(link.get('href'))
```

```python
#To extract only the table rows: first 10

rows = soup.find_all('tr')

print(rows[:10])
```

```python
#Main goal is to extract the data (table rows in this case) and converting them to dataframe
```

```python
rows = soup.find_all('tr')

for row in rows:
    row_td.append(row.find_all('td'))
print(row_td)
```

```python
#Now we need to remove the tags from each row element and get just text

#to do this we use beautiful soup object: get_text method

cleantext = BeautifulSoup(str(row_td), 'lxml').get_text()

print(type(cleantext))


print(cleantext)
```

```python
#Using regex to remove tags

import re

list_rows =[]

for row in rows:
    cell = row.find_all('td')
    str_cell = str(cell)
    
    #First, you compile a regular expression by passing a string to match to re.compile(). 
    #The dot, star, and question mark (.*?) will match an opening angle bracket followed by anything and 
    #followed by a closing angle bracket. It matches text in a non-greedy fashion, that is, 
    #it matches the shortest possible string. If you omit the question mark, it will match all the text 
    #between the first opening angle bracket and the last closing angle bracket. 
    #After compiling a regular expression, you can use the re.sub() method to find all the substrings 
    #where the regular expression matches and replace them with an empty string. 
    
    clean = re.compile('<.*?>')
    clean2 = (re.sub(clean, '', str_cell))
    list_rows.append(clean2)
print(list_rows)
type(list_rows)
```

```python
#converting above clean data list to DataFrame

df = pd.DataFrame(list_rows)

#displaying 10 rows only

df.head(10)
```

```python
#Now comes the data manipulation and cleaning

#Splitting the record by comma; currently there is one column '0'; so we will split it into multiple columns

df1 = df[0].str.split(',', expand = True)
df1.head(10)
```

```python
#Now removing the unwanted square brackets from the data frame on column 0 and 1

df1[0] = df1[0].str.strip('[')
df1[1] = df1[1].str.strip(']')

df1.head(10)
```

```python
#Adding headers to the dataframe above

#First fetch the header
clean_headers=[]
headers = soup.find_all('th')
#Extracting only the text using soup object

clean3 = BeautifulSoup(str(headers), 'lxml').get_text()
clean_headers.append(clean3) #creating a list of headers
print(clean_headers)
```

```python
#creating a dataframe from the list of headers

df2 = pd.DataFrame(clean_headers)
df2.head()
```

```python
#splitting column 0 into multiple columns 

df3 = df2[0].str.split(',', expand = True)
df3.head()
```

```python
#Merging these headers data frame df3 to original data frame containing table data df1

frames =[df3, df1]

df4 = pd.concat(frames)
df4.head(10)
```

```python
#making first row as header

df5 = df4.rename(columns=df4.iloc[0])
df5.head(10)
```

```python
#to get overview of the data 

df5.info() #table structure info
df.shape #gives no of rows and columns
```

```python
#dropping the first row since its already used as header and other irrelevant rows

df6 = df5.drop(df5.index[0:5])
df6.head()
```

```python
#Renaming column names to appropriate names

df6.rename(columns={'[Place': 'Place'}, inplace=True)
df6.rename(columns={' Team]':'Team'}, inplace=True)

df6.head()
```

```python
#Now final step is removing ']' from team column

df6['Team']=df6['Team'].str.strip(']')
df6.head()
```

```python
#Data analysis and Visualization

#creating a new column containing time in minutes to complete 10K run

time_list = df6[' Chip Time'].tolist()
min_list=[]
for i in time_list:
    h, m, s= i.split(':')
    #converting to minutes
    minutes = (int(h) * 3600 + int(m) * 60 + int(s))/60
    min_list.append(minutes)
print(min_list)
```
